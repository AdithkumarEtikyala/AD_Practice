```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
data1=pd.read_csv("NewspaperData.csv")
data1
```

```python
data1.info()
```

```python
data1.describe()
```

```python
plt.figure(figsize=(6,5))
boxplot_data = plt.boxplot(data1["daily"], vert=False)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data1['daily'], bins=30, kde=True)
plt.title('Histogram of daily')
plt.xlabel('daily')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data1['daily'], shade=True)
plt.title('KDE Plot of daily')
plt.xlabel('daily')
plt.ylabel('Density')
plt.show()
```

```python
plt.scatter(data1["daily"],data1["sunday"])
```

```python
import statsmodels.formula.api as smf
model=smf.ols("sunday~daily",data=data1).fit()
```
